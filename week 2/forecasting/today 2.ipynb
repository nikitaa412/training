```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load and preprocess data
df = pd.read_csv('weatherHistory.csv')
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True).dt.tz_convert(None)
df.set_index('Formatted Date', inplace=True)

# Filter numeric columns and resample
df = df[['Humidity', 'Wind Speed (km/h)', 'Temperature (C)']].dropna()
df.columns = ['humidity', 'windspeed', 'temperature']
df = df.resample('H').mean().dropna()

# Remove outliers
df = df[(df['temperature'] > -30) & (df['temperature'] < 50)]
df = df[(df['humidity'] >= 0) & (df['windspeed'] >= 0)]

# Scale features
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[['humidity', 'windspeed', 'temperature']])
df_scaled = pd.DataFrame(scaled_features, index=df.index, columns=['humidity', 'windspeed', 'temperature'])

# Create sequences
def create_sequences(data, target_index, time_steps=24):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i+time_steps])
        y.append(data[i + time_steps, target_index])  # Predicting temperature
    return np.array(X), np.array(y)

time_steps = 24
X, y = create_sequences(df_scaled.values, target_index=2, time_steps=time_steps)  # temperature is index 2

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Build LSTM regression model
model = Sequential()
model.add(LSTM(64, return_sequences=False, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # regression output

model.compile(loss='mean_squared_error', optimizer=Adam(0.001))

# Train
history = model.fit(X_train, y_train, epochs=15, batch_size=64, validation_split=0.1, verbose=1)

# Predict
y_pred = model.predict(X_test).flatten()

# Inverse scale temperature to original values
temp_scaler = MinMaxScaler()
temp_scaler.min_, temp_scaler.scale_ = scaler.min_[2], scaler.scale_[2]
y_test_rescaled = y_test * scaler.data_range_[2] + scaler.data_min_[2]
y_pred_rescaled = y_pred * scaler.data_range_[2] + scaler.data_min_[2]

# Evaluation metrics
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_rescaled, y_pred_rescaled)

print("\n--- Regression Metrics ---")
print(f"MAE:  {mae:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R:   {r2:.4f}")

# Plot actual vs predicted
plt.figure(figsize=(12, 5))
plt.plot(y_test_rescaled[:200], label='Actual')
plt.plot(y_pred_rescaled[:200], label='Predicted')
plt.title('Actual vs Predicted Temperature')
plt.xlabel('Time Step')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR
from datetime import timedelta

# Load and preprocess data
df = pd.read_csv('weatherHistory.csv')

# Parse datetime correctly and remove timezone (if present)
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['Formatted Date'] = df['Formatted Date'].dt.tz_convert(None)

# Select relevant columns
df_var = df[['Formatted Date', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']]
df_var.set_index('Formatted Date', inplace=True)

# Resample data daily to remove noise
df_daily = df_var.resample('D').mean().dropna()

# Train VAR model
model = VAR(df_daily)
results = model.fit(maxlags=15, ic='aic')

# Forecast steps: days until 2026-12-31
last_date = df_daily.index[-1]
end_date = pd.Timestamp('2026-12-31')
forecast_days = (end_date - last_date).days

# Forecast future values
forecast = results.forecast(df_daily.values[-results.k_ar:], steps=forecast_days)
forecast_index = [last_date + timedelta(days=i + 1) for i in range(forecast_days)]
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=df_daily.columns)

# Save the forecasted data to a .h5 file
forecast_df.to_hdf('weather_forecast.h5', key='forecast', mode='w')

# Plot Temperature Forecast
plt.figure(figsize=(15, 6))
plt.plot(df_daily['Temperature (C)'], label='Historical Temperature')
plt.plot(forecast_df['Temperature (C)'], label='Forecast Temperature', linestyle='--')
plt.title('Temperature Forecast using VAR')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()

# Plot Humidity Forecast
plt.figure(figsize=(15, 6))
plt.plot(df_daily['Humidity'], label='Historical Humidity')
plt.plot(forecast_df['Humidity'], label='Forecast Humidity', linestyle='--')
plt.title('Humidity Forecast using VAR')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.legend()
plt.grid(True)
plt.show()

# Plot Wind Speed Forecast
plt.figure(figsize=(15, 6))
plt.plot(df_daily['Wind Speed (km/h)'], label='Historical Wind Speed')
plt.plot(forecast_df['Wind Speed (km/h)'], label='Forecast Wind Speed', linestyle='--')
plt.title('Wind Speed Forecast using VAR')
plt.xlabel('Date')
plt.ylabel('Wind Speed (km/h)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Load and preprocess data
df = pd.read_csv('weatherHistory.csv')
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
df['Formatted Date'] = df['Formatted Date'].apply(lambda x: x.replace(tzinfo=None))  # Remove timezone

# Prepare data for Prophet
humidity_df = df[['Formatted Date', 'Humidity']].rename(columns={'Formatted Date': 'ds', 'Humidity': 'y'})

# Build and fit the model
humidity_model = Prophet()
humidity_model.fit(humidity_df)

# Create future dates till 2026
years_to_predict = 2026 - humidity_df['ds'].dt.year.max() + 1
future = humidity_model.make_future_dataframe(periods=365 * years_to_predict)
forecast = humidity_model.predict(future)

# Plot
humidity_model.plot(forecast)
plt.title('Humidity Forecast till 2026')
plt.xlabel('Date')
plt.ylabel('Humidity')
plt.show()
```
