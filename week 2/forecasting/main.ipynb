```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('weatherHistory 1.csv')

# Convert "Formatted Date" to datetime
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True, errors='coerce')

# List of continuous columns to plot
continuous_columns = [
    "Temperature (C)",
    "Apparent Temperature (C)",
    "Humidity",
    "Wind Speed (km/h)",
    "Wind Bearing (degrees)",
    "Visibility (km)",
    "Loud Cover",
    "Pressure (millibars)"
]

# Plot each continuous variable against time
plt.figure(figsize=(16, 24))
for i, col in enumerate(continuous_columns, 1):
    plt.subplot(4, 2, i)
    plt.plot(df['Formatted Date'], df[col], label=col, linewidth=0.5)
    plt.xlabel('Time')
    plt.ylabel(col)
    plt.title(f'{col} over Time')
    plt.tight_layout()

plt.show()
```
