```python
import pandas as pd
import numpy as np
from scipy import stats
from IPython.display import display
```

```python
csv_path = (r"C:\Users\NikitaRampurkarV-Sof\nikita_vsoft\tasks\Data Analysis\data preprocessing\diabetes.csv")  # Change this to your actual file path
df = pd.read_csv(csv_path)
```

```python
num_df = df.select_dtypes(include='number')
```

```python
stats_df = pd.DataFrame(index=num_df.columns)
```

```python
stats_df['Mean'] = num_df.mean()
stats_df['Median'] = num_df.median()
stats_df['Mode'] = num_df.mode().iloc[0]  # First mode
stats_df['Min'] = num_df.min()
stats_df['Max'] = num_df.max()
stats_df['Standard Deviation'] = num_df.std()
stats_df['Variance'] = num_df.var()
stats_df['Standard Error'] = num_df.sem()
stats_df['25th Percentile'] = num_df.quantile(0.25)
stats_df['50th Percentile (Median)'] = num_df.quantile(0.5)
stats_df['75th Percentile'] = num_df.quantile(0.75)
stats_df['Skewness'] = num_df.skew()
stats_df['Kurtosis'] = num_df.kurt()
```

```python
print("ðŸ“Š Descriptive Statistics for Numeric Columns:")
display(stats_df.round(4))
```

```python
# --- One-Way ANOVA: Between each pair of numeric columns ---
anova_results = {}
columns = num_df.columns
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        f_stat, p_val = stats.f_oneway(num_df[columns[i]], num_df[columns[j]])
        anova_results[f"{columns[i]} vs {columns[j]}"] = {
            'F-statistic': f_stat,
            'p-value': p_val
        }

anova_df = pd.DataFrame(anova_results).T
```

```python
# Display ANOVA Results
print("\nðŸ“ˆ One-Way ANOVA Results (pairwise):")
display(anova_df.round(4))
```
